### Simple charts to understand loss



## Import necessary lib

```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Define the function to create figure 

```python
def create_figure():
    fig = plt.figure(figsize=(8,6), dpi=72,facecolor="white")
    axes = fig.add_subplot(111)

    axes.spines['top'].set_color('none')
    axes.spines['right'].set_color('none')

    axes.xaxis.set_ticks_position('bottom')
    axes.spines['bottom'].set_position(('data',0))
    axes.yaxis.set_ticks_position('left')
    axes.spines['left'].set_position(('data',0))
    
    return fig,axes
```

```python
a = 3
b = 4
```

## Display ax+b

```python

fig,axes = create_figure()

data_x = [n for n in range(-2, 3)]
data_y = [n for n in range(-2, 3)]

for i in range(0,len(data_x)):
    data_y[i] = data_x[i]*a + b + random.uniform(0,1) 

    
axes.scatter(data_x, data_y, marker = 'o', color = 'b', label='3', s = 15) 

x = np.linspace(-3, 3)
y = a*x + b

axes.plot(x,y, color = 'red', linewidth=2, linestyle="-")


axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

## When we change the value of b

```python
fig,axes = create_figure()

axes.scatter(data_x, data_y, marker = 'o', color = 'b', label='3', s = 15) 


for i in range(-3, 4):
    x = np.linspace(-3, 3)
    y = a*x + (b + i*2)
    if (i == 0):
        axes.plot(x,y, color = 'red', linewidth=2, linestyle="-")
    else:
        axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")

line_y = np.linspace(0, 20)
line_x = np.linspace(2, 2)

axes.plot(line_x,line_y, color = 'black', linewidth=2, linestyle="-")


#axes.plot(x,y, color = 'red', linewidth=2, linestyle="-")


axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

```python
fig,axes = create_figure()

loss_x = np.linspace(-5,13)
loss_y = np.square(b - loss_x)

axes.plot(loss_x,loss_y, color = 'black', linewidth=2, linestyle="-")

axes.set_xlim(1.1*loss_x.min(),1.1*loss_x.max())
axes.set_ylim(-10,1.1*loss_y.max())

plt.show() 
```

```python
fig,axes = create_figure()

axes.scatter(data_x, data_y, marker = 'o', color = 'b', label='3', s = 15) 

for i in range(-3, 4):
    x = np.linspace(-3, 3)
    y = (a+i)*x + b 
    if (i == 0):
        axes.plot(x,y, color = 'red', linewidth=2, linestyle="-")
    else:
        axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")
    
line_y = np.linspace(0, 20)
line_x = np.linspace(2, 2)

axes.plot(line_x,line_y, color = 'black', linewidth=2, linestyle="-")


axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

```python
fig,axes = create_figure()

loss_x = np.linspace(-3,9)
loss_y = np.square(a*2 - loss_x*2)

axes.plot(loss_x,loss_y, color = 'black', linewidth=2, linestyle="-")

axes.set_xlim(1.1*loss_x.min(),1.1*loss_x.max())
axes.set_ylim(-10,1.1*loss_y.max())

plt.show() 
```
