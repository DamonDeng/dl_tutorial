### Simple python code to understand Sigmoid

The sigmoid activation function is:

f(x) = 1/(1+1/exp(x))

The following python code will show you what it looks like in 2D chart.

## Import necessary lib

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Define the function to create figure 

```python
def create_figure():
    fig = plt.figure(figsize=(8,6), dpi=72,facecolor="white")
    axes = fig.add_subplot(111)

    axes.spines['top'].set_color('none')
    axes.spines['right'].set_color('none')

    axes.xaxis.set_ticks_position('bottom')
    axes.spines['bottom'].set_position(('data',0))
    axes.yaxis.set_ticks_position('left')
    axes.spines['left'].set_position(('data',0))
    
    return fig,axes
```

```python
a = 3
b = 4
```

## Display ax+b

```python

fig,axes = create_figure()
    
x = np.linspace(-3, 3)
y = a*x + b

axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")

axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

## Display exp(ax+b)

```python
fig,axes = create_figure()
    
x = np.linspace(-3, 3)
y = np.exp(a*x+b)

axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")

axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

## Display 1/exp(ax+b)

```python
fig,axes = create_figure()
    
x = np.linspace(-3, 3)
y = 1/np.exp(a*x+b)

axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")

axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

## Display 1+1/exp(ax+b)

```python
fig,axes = create_figure()
    
x = np.linspace(-3, 3)
y = 1+1/np.exp(a*x+b)

axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")

axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-10,1.1*y.max())

plt.show() 
```

## Display 1/(1+1/exp(ax+b))

```python
fig,axes = create_figure()
    
x = np.linspace(-3, 3)
y = 1/(1+1/np.exp(a*x+b))

axes.plot(x,y, color = 'blue', linewidth=2, linestyle="-")

axes.set_xlim(1.1*x.min(),1.1*x.max())
axes.set_ylim(-0.5,1.1*y.max())

plt.show() 
```
